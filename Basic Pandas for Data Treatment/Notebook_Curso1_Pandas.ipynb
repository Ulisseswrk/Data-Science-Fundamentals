### Exercício 3 - Pandas Alura

1- Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

```python
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
dados.fillna(0, inplace = True)
dados
```

2- Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

```python
dados.drop(dados.query('Nome == "Alice" | Nome == "Carlos"').index, inplace = True)
dados
```

3- Aplique um filtro que selecione apenas os alunos que foram aprovados e modifique as notas 7 para 8.

```python
aprovados = dados.query('Aprovado == True').reset_index(drop=True)
aprovados.replace(7.0,8.0)
```

4- Crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles. (Continuação do Exercício)

```python
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x * 0.4)
```

5- Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados
```

6- Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

Sim: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
Não: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x: 'Sim' if x >= 6 else 'Não')
dados['Notas_finais'] = dados['Notas_finais'].apply(lambda x: 10 if x > 10 else x) # Código impedindo o aluno de ficar com a nota acima de 10
dados
```

7- Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
alunos_salvos = dados.query('Aprovado == False & Aprovado_final == "Sim"')
```

8- Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
dados.to_csv('dados_alunos.csv', index = False, sep = ';')
```

# Manipulando os dados

## Criando colunas numéricas

```python
df['Valor_por_mes'] = df['Valor'] + df['Condominio'] # Criando coluna de gastos mensais com os imóveis
```

```python
df['Valor_por_ano'] = df['Valor_por_mes'] * 12 + df['IPTU'] # Criando coluna de gastos anuais com os imóveis
df.head()
```

## Criando colunas categóricas

```python
#Adicionamos uma nova coluna chamada descrição com todas as informações, mas pra concatenar str com int ou float usasse o astype(str)
df['Descricao'] = df['Tipo'] + ' em ' + df['Bairro'] + ' com ' + \
                                        df['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + df['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
# Função que analisa se tem suite e retorna sim ou não pra coluna possui suite
df['Possui_suite'] = df['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
df.head(10)
```
